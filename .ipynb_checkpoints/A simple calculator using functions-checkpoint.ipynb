```python
def menu():
    print("*******************CALCULATOR*******************")
    validChoices = ["A", "S", "M", "D", "P2", "P3"]
    choice = input ("""
                        For Addition: A
                        For Subtraction: S
                        For Multiplication: M
                        For Division: D
                        For One Number to the Power of 2: P2
                        For One Number to the Power of 3: P3

                        Please enter your choice:  """)
    
    if choice not in validChoices:
        choice = input("Try again: ")

    return choice

def enterNumber():
    valid = False
    while not valid:
        try:
            x = int(input("\nEnter number: "))
            valid = True
        except ValueError:
            x = int(input("\nTry again: "))
    return x


def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    return x / y

def power2(x):
    return x * x

def power3(x):
    return x * x * x

#main program starts here
print("Welcome to my Calculator")
run = True
done = False
userChoice = menu()
while run == True:
    if userChoice == "A":
        answer = add(enterNumber(), enterNumber())
        done = True
    elif userChoice == "S":
        answer = subtract(enterNumber(), enterNumber())
        done = True
    elif userChoice == "M":
        answer = multiply(enterNumber(), enterNumber())
        done = True
    elif userChoice == "D":
        answer = divide(enterNumber(), enterNumber())
        done = True
    elif userChoice == "P2":
        answer = power2(enterNumber())
        done = True
    elif userChoice == "P3":
        answer = power3(enterNumber())
        done = True
    print("\nThe answer from your calculation is = "+str(answer))
    if done == True:
        menu()
        
```
